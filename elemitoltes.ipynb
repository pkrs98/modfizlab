## Elemi töltés meghatározása 
### (Fizika labor 3 - 2. mérés)

```python
# szükséges importok
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# függvény a laborjegyzetben felírt képlet implementálására
def f(x,q):
    fx = np.zeros((1,len(x)))
    for j in range(0,len(q)-1):
        a = []
        for i in range(0,len(x)):
            a1 = math.sin(math.pi*(q[j]/x[i]))
            a2 = a1**2
            a = np.append(a, a2)
        fx += a
    return fx
```

```python
# adatok és x tartomány:
x = np.linspace(-1.1E-30,1.75E-19,1000)
qk = [-1.024E-19, 2.479E-19, 3.715E-19, -1.139E-18, 5.13E-19,
     1.335E-19, 1.654E-19, 1.927E-19, 4.229E-19, -1.25E-19,
     -3.716E-20, -1.564E-19, -3.697E-20, 2.421E-20, 6.602E-20,
      1.614E-19, 1.513E-19, 1.06E-19, 1.057E-19, -1.71E-20,
      6.222E-20, 6.751E-20, 4.259E-20, 1.949E-19]
```

```python
fx = f(x,qk)
fx = fx[0,:]
```

```python
# kiszámolt függvény plottolása
plt.plot(x,fx,label="f(x)")
plt.xlabel('x [C]')
plt.ylabel('y [C]')
plt.title('Elemi töltés meghatározásához szükséges függvény alakja')
plt.grid()
plt.legend()
```

A vártaknak megfelelően nulla közelében a függvényben zaj figyelhető meg, ami körülbelül x=1,23*10^-19 körül cseng le. 

Innen nézzük a lokális minimumokat, és keressük közülük a legnagyobbat.

```python
# legnagyobb lokális minimum keresése x>1,5*10^-19-től
slic_e = fx[ 1.6E-19 < x ]
MaxLokMin = x [ fx == min(slic_e)]
MaxLokMin_y = fx [ x == MaxLokMin ] 
```

```python
# ábrázoljuk a lényeges tartományt, és az elemi töltés helyét
plt.plot(x[700:],fx[700:],label="f(x)")
plt.plot(MaxLokMin, MaxLokMin_y, "o", c="black", label="elemi töltés")
plt.title("Elemi töltés meghatározása")
plt.xlabel("x [C]")
plt.ylabel("y [C]")
plt.grid()
plt.legend()
```

```python
print("Az általunk mért olajcseppek töltéséből meghatározott elemi töltés: ",MaxLokMin," C.")
```
