```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def delta(x1,y1,x):
    y = []
    for i in range(len(x)):
        x1p = x1+0.25
        x1m = x1-0.25
        if x1p > x[i] > x1m:
            y = np.append(y,y1)
        else:
            y = np.append(y,0)
    return y
```

```python
x = np.linspace(30,155,10000)
x1s = [33.8, 48.6, 60.6, 71.2, 81.3, 91.0, 111.0, 121.8, 134.2, 150.2]
y1s = [0.33,0.33,1,1,0.33,0.33,1,0.33,0.33,1]
```

```python
ys = []
for i in range(len(x1s)):
    ys = np.append(ys,delta(x1s[i],y1s[i],x))
a = 10000
```

```python
plt.plot(x,ys[:a],lw=2,label="100, N=1")
plt.plot(x,ys[a:2*a],lw=2,label="110, N=2")
plt.plot(x,ys[2*a:3*a],lw=2, label="111, N=3")
plt.plot(x,ys[3*a:4*a],lw=2,label="200, N=4")
plt.plot(x,ys[4*a:5*a],lw=2, label="210, N=5")
plt.plot(x,ys[5*a:6*a], lw=2, label="211, N=6")
plt.plot(x,ys[6*a:7*a], lw=2, label="220, N=8")
plt.plot(x,ys[7*a:8*a], lw=2, label="300 221, N=9")
plt.plot(x,ys[8*a:9*a], lw=2, label="301, N=10")
plt.plot(x,ys[9*a:10*a])
plt.grid()
plt.xlabel("2 theta [°]")
plt.ylabel("Relatív intenzitás")
plt.title("Adott kristályos anyag pordiffrakciójának eredménye, \n1e-10 m hullámhosszúságú besugárzással")
plt.legend(loc=(1.04,0))
```
